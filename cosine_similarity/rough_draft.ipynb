# make function to solve missing data

```python
def rename(row):
    if row[ds.columns[2]]=="Unknown" :
        return row[ds.columns[3]]          
    elif row[ds.columns[2]]!="Unknown":
        return row[ds.columns[2]]
    elif row[ds.columns[3]]=="Unknown":
        return row[ds.columns[2]]
```

```python
ds['name']= ds.apply(lambda row: rename(row), axis=1)
```

```python
ds.head(50)
```

```python
ds=ds.drop(ds.columns[[1,2,3,4,9]], axis=1)
```

```python
import scipy as sp #pivot egineering
import scipy.sparse
pivot = ds.pivot_table(index=['name'], columns=ds[ds.columns[0]], values='Score')
pivot.fillna(0, inplace=True)
pivot.head()
```

```python
pivot.shape
```

```python
from scipy.sparse import csr_matrix
pivot_sparse = csr_matrix(pivot)
```

```python
#ML model
from sklearn.metrics.pairwise import  linear_kernel , cosine_similarity
```

# make model

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
model.fit(pivot_sparse)
```

```python
distances, suggestions = model.kneighbors(pivot.iloc[10, :].values.reshape(1, -1))
```

# test model

```python
for i in range(len(suggestions)):
  print(pivot.index[suggestions[2]])
```
